```python
import pandas as pd
import numpy as np
import re
import os
```

```python
fileName = 'XXXXXXXXX'
```

```python
currpath = os.getcwd() + "/soft_files/" + fileName
file = open(currpath).read().split("\n")
c=0
matrix = []
identifiers = []
meta_data = []
marked = False

for line in file:
    if not marked and re.search(r'!dataset_table_begin', line):
        marked = True
        continue
        
    if marked:
        if re.search(r'!dataset_table_end' , line):
            break
        
        
        line = line.split("\t")
        if c == 0:
            for ident in line:
                identifiers.append(ident)
            c=+1
            continue
        
        temp = []
        for num in line:
            t = num.lower()
            if str(num) == "":
                continue
            if re.search("control", t):
                temp.append("Control")
            else:
                temp.append(num)

        matrix.append(temp)
```

```python
df = pd.DataFrame(matrix, columns=identifiers)
df
```

```python

to = "/csv_files/" + str(fileName) + '.csv'
path = os.getcwd() + to
df.to_csv(path, index=False, header=True)
```
